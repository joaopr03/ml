```python
train_data['Accident Date'] = pd.to_datetime(train_data['Accident Date'])
```

```python
accidents_per_month = train_data.groupby(train_data['Accident Date'].dt.to_period('M')).size()
```

```python
# Plot the bar chart
accidents_per_month.plot(kind='bar', figsize=(6,4), color='skyblue')
plt.title('Number of Accidents per Month')
plt.xlabel('Accident Date')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Define age categories
bins = [0, 17, 30, 40, 50, 60, 74, 100]
labels = ['0-17', '18-30', '31-40', '41-50', '51-60', '61-74', '75+']

# Create a new column for age categories
train_data['Age Category'] = pd.cut(train_data['Age at Injury'], bins=bins, labels=labels, right=False)
```

```python
# Count the number of occurrences in each age category
age_distribution = train_data['Age Category'].value_counts()
```

```python
# Plot the pie chart
plt.figure(figsize=(7, 7))  # Set the figure size

age_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'coral', 'orange', 'purple', 'yellow', 'pink'])

# Customize the plot
plt.title('Age Distribution of Accidents')
plt.ylabel('')  # Remove the default y-label (which shows "Age Category")

# Show the plot
plt.show()
```

```python
train_data['Day of Year'] = train_data['Accident Date'].dt.dayofyear
```

```python
# Map Gender to numerical values (e.g., 0 for Male, 1 for Female)
train_data['Gender Numeric'] = train_data['Gender'].map({'M': 0, 'F': 1})
```

```python
# Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(train_data['Day of Year'], train_data['Gender Numeric'], alpha=0.6, color='skyblue')

# Add labels and title
plt.title('Scatter Plot of Accident Date vs Gender', fontsize=14)
plt.xlabel('Day of the Year (Accident Date)', fontsize=12)
plt.ylabel('Gender (0=M, 1=F)', fontsize=12)

# Customizing the y-ticks for better interpretation
plt.yticks([0, 1], ['M', 'F'])

# Show the plot
plt.show()
```

```python
# Step 2: Loop through each column (variable) and create a histogram
for i in range(len(train_data.columns)):
    column_name = train_data.columns[i]  # Get the column name
    plt.figure(figsize=(8, 5))  # Optional: Set figure size for better visuals
    plt.hist(train_data[column_name].dropna(), bins=30, edgecolor='k', alpha=0.7)  # Drop NaNs and create the histogram
    plt.title(f'Histogram of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.grid(True)  # Optional: Add grid
    plt.show()  # Display the plot
```
